```python
import gph
import numpy as np

points = np.random.randint(20, size=(20, 2))
points
```

```python
output = gph.ripser_parallel(points, return_generators=True)
output
```

```python
print("Persistance pairs: ")
print("H_0: ", output['dgms'][0])
print("H_1: ", output['dgms'][1])
```

```python
print(output['gens'][0])
print()
print(output['gens'][1])
print()
print(output['gens'][2])
print()
print(output['gens'][3])
```

```python
print("Generators for corresponding pairs: ")
print("H_0: ")
for i in range(len(output['gens'][0])):
  print(output['gens'][0][i][0], ' ---> ', output['gens'][0][i][1:])
print("first number is the 'birth' vertex, second and third number are vertices of the edge that `killed' the topological feature for all FINITE bars in H_0 barcode\n\n")
for i in range(len(output['gens'][2])):
  print(output['gens'][2][i], ' ---> +infty')
print("first number is the 'birth' vertex for all IN_FINITE bars in H_0 barcode")


print("\n ----------------------------")

for i in range(len(output['gens'][1][0])):
  print(output['gens'][1][0][i][:2], ' ---> ', output['gens'][1][0][i][2:])
print("first two numbers are  vertices of the edge that 'created' topological feature, third and forth numbers are vertices of the edge that `killed' the topological feature for all FINITE bars in H_1 barcode")
for i in range(len(output['gens'][3])):
  print(output['gens'][3][i], ' ---> +infty')
print("first two numbers are  vertices of the edge that 'created' topological feature for all IN_FINITE bars in H_1 barcode")
```

```python
import numpy as np

H0_birth = np.zeros(points.shape[0], dtype=int)
H0_death = np.zeros((points.shape[0], 2), dtype=int)

H1_birth = np.zeros((points.shape[0], 2), dtype=int)
H1_death = np.zeros((points.shape[0], 2), dtype=int)

# Iterate over finite generators for H0
for i in range(len(output['gens'][0])):
    birth_vertex = output['gens'][0][i][0]
    death_vertices = output['gens'][0][i][1:]

    H0_birth[birth_vertex] = 1
    H0_death[death_vertices[0], 0] = 1
    H0_death[death_vertices[1], 1] = 1

# Iterate over finite generators for H0
for i in range(len(output['gens'][2])):
    birth_vertex = output['gens'][2][i]

    H0_birth[birth_vertex] = 1

# Iterate over finite generators for H1
for i in range(len(output['gens'][1][0][:2])):
    birth_vertices = output['gens'][1][0][i][:2]
    death_vertices = output['gens'][1][0][i][2:]

    H1_birth[birth_vertices[0], 0] = 1
    H1_birth[birth_vertices[1], 1] = 1
    H1_death[death_vertices[0], 0] = 1
    H1_death[death_vertices[1], 1] = 1

# Iterate over finite generators for H1
for i in range(len(output['gens'][3])):
    birth_vertex = output['gens'][3][i]

    H1_birth[birth_vertices[0], 0] = 1
    H1_birth[birth_vertices[0], 0] = 1
```

```python
import pandas as pd

# Assuming x_coords and y_coords are arrays containing x and y coordinates respectively
data = {'x': points[:, 0], 'y': points[:, 1]}

# Convert H0_birth, H0_death, H1_birth, H1_death to DataFrame
df_H0_birth = pd.DataFrame(H0_birth, columns=['H0_birth'])
df_H0_death = pd.DataFrame(H0_death, columns=['H0_death_1', 'H0_death_2'])
df_H1_birth = pd.DataFrame(H1_birth, columns=['H1_birth_1', 'H1_birth_2'])
df_H1_death = pd.DataFrame(H1_death, columns=['H1_death_1', 'H1_death_2'])

# Concatenate all DataFrames
df = pd.concat([pd.DataFrame(data), df_H0_birth, df_H0_death, df_H1_birth, df_H1_death], axis=1)

# Display the DataFrame
df
```
